# Subplots

plt.subplots will return two objects: a figure object, and a tuple of axes objects

figure object: overall figure canvas on which the axes are overlaid
axes object: individual subplot canvases

plt.subplots() will accept a number of keyword arguments, the most important being
figsize: (width, height)
nrows: (integer) the number of rows
ncols: (integer) the number of columns

from now on -> anywhere that you were going to say plt. , instead say ax.
also, instead of just .title or .ylim, etc. etc. you do .set_title, .set_ylim, etc. etc.

```python
fig, ax = plt.subplots(figsize=(16, 9), nrows=2, ncols=2)
ax[0, 0].plot(df.l, df.v)
ax[0, 0].set_title('v vs. l')
ax[0, 0].set_ylabel('v')
ax[0, 0].set_xlabel('l')

sns.lineplot(data=df, x='l', y='v', ax=ax[0, 1])
ax[0, 1].set_title('v vs. l using seaborn')
```
