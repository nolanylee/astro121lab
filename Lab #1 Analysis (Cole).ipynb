```python
import ugradio #if remote comment this section out
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
import pandas as pd
```

```python
df = pd.read_csv('test_data_012222')
x = np.arange(2048)
```

```python
data_100_230 = df['fgen_100_sample_230']
plt.figure()
plt.plot(x, data_100_230)
```

```python
four_100_230 = ugradio.dft.dft(data_100_230)
plt.figure()
plt.title('Fgen_100_Sample_230 DFT')
plt.plot(four_100_230[0], four_100_230[1].real, label = 'Real')
plt.plot(four_100_230[0], four_100_230[1].imag, label = 'Imaginary')
plt.legend()
```
