```python
import ugradio
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import new_interf
import seaborn as sns
from scipy.fft import fft
from scipy.signal import butter, lfilter, freqz

from datetime import datetime, timedelta

sns.set_theme(style='ticks', palette='magma')
```

```python
def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y
order = 3
fs = 10
cutoff = 0.001

df = pd.read_csv('hour_data.csv')
df.drop('Unnamed: 0', inplace=True, axis=1)
dt = df.all_times[1]-df.all_times[0]
df['volt_spec'] = np.fft.fft(np.array(df.all_voltages))

df.loc[0, 'volt_spec'] = 0
df['power_spec'] = np.abs(df['volt_spec'])**2
df['frequencies'] = np.fft.fftfreq(len(df), dt)
df['Shifted Frequency'] = np.fft.fftshift(df['frequencies'])
df['Shifted Power Spectrum'] = np.fft.fftshift(df['power_spec'])
df['ifft'] = np.abs(np.fft.ifft(df['volt_spec']))
df['scaled_times'] = df.all_times-df.all_times[0]
df['buttered_ifft']=np.abs(butter_lowpass_filter(df['volt_spec'], cutoff, fs, order))
df['Voltage Spectrum']=np.abs(df['volt_spec'])
df
```

```python
fig, axes = plt.subplots()
sns.lineplot(data=df, x='scaled_times', y='buttered_ifft', ax=axes, label='buttered')
sns.lineplot(data=df, x='scaled_times', y='ifft', ax=axes, label='ifft')
```
